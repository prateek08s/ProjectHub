```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import plotly.express as px
```

```python
hk_project = reviews_all('com.hikingproject.android', sleep_milliseconds=0, lang='en',country='US', sort=Sort.NEWEST)
```

```python
#hk_project
```

```python
df = pd.json_normalize(hk_project)
df.head()
```

```python
df['score'].mean()
```

```python
df['reviewCreatedVersion'].value_counts()
```

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")

```

```python
print(sentiment_analysis("I love this!"))
```

```python
df.dtypes
```

```python
df['content'] = df['content'].astype('str')
```

```python
df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
df.head()
```

```python
df.shape
```

```python
y=sentiment_analysis("I love this!")
```

```python
y[0]['label']
```

```python
y[0]['score']
```

```python
df['sentiment']=df['result'].apply(lambda x: (x[0]['label']))
df['score']=df['result'].apply(lambda x: (x[0]['score']))
```

```python
df['score'].mean()
```

```python
df['sentiment'].value_counts(normalize=True) #(normalize=True) will give it in percentage
```
